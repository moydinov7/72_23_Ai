```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Yangi ma'lumotlar: yosh va bo'y
mahsulotlar = {
    'yosh': [21, 30, 25, 55, 33, 43, 22, 45, 29, 52, 26, 38, 27, 23, 31, 40, 24, 35, 36, 48, 20, 28, 50, 19, 41, 47, 32, 37, 34, 46],
    'boy': [162, 175, 180, 170, 185, 178, 174, 176, 179, 165, 182, 180, 177, 170, 183, 181, 173, 186, 187, 169, 160, 178, 166, 161, 182, 168, 184, 185, 187, 170]
}
```

```python
# DataFrame yaratish
df = pd.DataFrame(mahsulotlar)
```

```python
# Kirish va chiqish o'zgaruvchilar
X = df[['yosh']]
y = df['boy']
```

```python
# Train/Test bo'lish
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# Model yaratish va o'qitish
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Og‘irlik va intercept
ogirlik = model.coef_[0]
bias = model.intercept_
print("Og‘irlik:", ogirlik)
print("Bias:", bias)
```

```python
# Bashorat chizig‘i
chiziq = ogirlik * X + bias
```

```python
# Vizualizatsiya
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Asl ma\'lumotlar')
plt.plot(X, chiziq, color='red', label='Chiziqli model')
plt.xlabel('Yosh')
plt.ylabel('Bo\'y')
plt.title('Yosh va Bo\'y o\'rtasidagi regressiya')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Test to'plamida bashorat
y_pred = model.predict(X_test)
```

```python
# Baholash
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R2: {r2:.2f}")
```

```python
# Maxsus bashorat
yosh_bashorat = 50
boy_bashorat = model.predict([[yosh_bashorat]])
print(f"{yosh_bashorat} yoshdagi inson uchun bashorat qilingan bo'y: {boy_bashorat[0]:.2f} sm")
```
