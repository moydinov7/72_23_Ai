```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor, plot_tree

# Daraxt 1: Oddiy
X1 = np.array([[70], [80], [85], [90], [95], [100]])
y1 = np.array([15000, 20000, 25000, 45000, 50000, 60000])
model1 = DecisionTreeRegressor(max_depth=3)
model1.fit(X1, y1)

# Daraxt 2: Kattaroq (yangi sun’iy ma’lumotlar bilan)
X2 = np.array([[60], [65], [75], [85], [95], [105], [115], [125]])
y2 = np.array([10000, 15000, 20000, 30000, 40000, 50000, 60000, 70000])
model2 = DecisionTreeRegressor(max_depth=4)
model2.fit(X2, y2)

# Daraxt 3: Eng katta daraxt (yanada ko‘proq ma’lumotlar bilan)
X3 = np.array([[50], [60], [70], [80], [90], [100], [110], [120], [130], [140]])
y3 = np.array([9000, 12000, 18000, 25000, 32000, 42000, 53000, 65000, 77000, 90000])
model3 = DecisionTreeRegressor(max_depth=5)
model3.fit(X3, y3)

# Daraxtlarni chizish
fig, axs = plt.subplots(1, 3, figsize=(24, 6))

plot_tree(model1, filled=True, feature_names=["House Size"], ax=axs[0], rounded=True)
axs[0].set_title("Decision Tree 1")

plot_tree(model2, filled=True, feature_names=["House Size"], ax=axs[1], rounded=True)
axs[1].set_title("Decision Tree 2")

plot_tree(model3, filled=True, feature_names=["House Size"], ax=axs[2], rounded=True)
axs[2].set_title("Decision Tree 3")

plt.tight_layout()
plt.show()
```
